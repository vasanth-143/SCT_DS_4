```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "dft-road-casualty-statistics-casualty-2023.csv"
df = pd.read_csv(file_path)

sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='casualty_severity', data=df)
plt.title("Casualty Severity Distribution")
plt.xlabel("Casualty Severity (1=Fatal, 2=Serious, 3=Slight)")
plt.ylabel("Number of Casualties")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='sex_of_casualty', data=df)
plt.title("Casualties by Sex")
plt.xlabel("Sex of Casualty (1=Male, 2=Female, 3=Unknown)")
plt.ylabel("Number of Casualties")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='age_band_of_casualty', data=df)
plt.title("Casualties by Age Band")
plt.xlabel("Age Band Code")
plt.ylabel("Number of Casualties")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
top_types = df['casualty_type'].value_counts().nlargest(10)
sns.barplot(x=top_types.index.astype(str), y=top_types.values)
plt.title("Top 10 Casualty Types")
plt.xlabel("Casualty Type Code")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
